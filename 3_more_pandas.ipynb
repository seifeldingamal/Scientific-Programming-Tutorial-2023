## 1. Data Cleaning

Load a dataset into a pandas DataFrame and clean the data by removing missing values, duplicates, and outliers.

```python
import pandas as pd
import numpy as np
```

Data Source and documentation: [Here.](https://www.kaggle.com/datasets/adityakadiwal/water-potability)

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('./assets/water_potability.csv')
df
```

```python
# Remove missing values
df.dropna(inplace=True)
df.reset_index(drop= True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
df.reset_index(drop= True)
```

```python
# Get Quantiles to identify outliers
Q1, Q3 = np.percentile(df['Solids'], q=[25, 75])
IQR = Q3 - Q1
Lower_Bound = Q1 - 1.5*IQR
Upper_Bound = Q3 + 1.5*IQR
print(f'Lower Bound = {Lower_Bound}' )
print(f'Upper Bound = {Upper_Bound}' )
```

```python
# Remove outliers
df[(df['Solids'] > Lower_Bound) & (df['Solids'] < Upper_Bound)]
```

## 2. Data Manipulation

Use pandas to manipulate the data by filtering rows based on conditions, sorting the data, and grouping the data by categories.

## 3. Data Visualization

 Use pandas and matplotlib to create visualizations of the data such as scatter plots, histograms, and box plots.

```python
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('./assets/water_potability.csv')
df
```

```python
# Create a scatter plot
plt.scatter(df['Solids'], df['Organic_carbon'])
plt.xlabel('Solids')
plt.ylabel('Organic carbon')
plt.title('Scatter Plot')
plt.show()
```

```python
# Create a histogram
plt.hist(df['Hardness'], bins=10)
plt.xlabel('Hardness')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Create a box plot
plt.boxplot(df['Chloramines'])
plt.xlabel('Chloramines')
plt.ylabel('Amount of Chloramines in ppm')
plt.title('Box Plot')
plt.show()
```
